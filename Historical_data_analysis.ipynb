```python
import pandas as pd

# Assuming df is your DataFrame containing flight data

# Step 1: Filter the DataFrame for the specific flight number
flight_number = int(input('Enter the flight number:'))
filtered_df = df_all_years[df_all_years['FL_NUMBER'] == flight_number]

# Step 2: Calculate departure and arrival delays
filtered_df['DEP_DELAY'] = filtered_df['DEP_TIME'] - filtered_df['CRS_DEP_TIME']
filtered_df['ARR_DELAY'] = filtered_df['ARR_TIME'] - filtered_df['CRS_ARR_TIME']

# Step 3: Evaluate on-time performance
total_flights = len(filtered_df)
on_time_departures = len(filtered_df[filtered_df['DEP_DELAY'] <= 0])
on_time_arrivals = len(filtered_df[filtered_df['ARR_DELAY'] <= 0])

departure_on_time_percentage = (on_time_departures / total_flights) * 100
arrival_on_time_percentage = (on_time_arrivals / total_flights) * 100

print(f"Flight Number: {flight_number}")
print(f"Total Flights: {total_flights}")
print(f"On-Time Departures: {on_time_departures} ({departure_on_time_percentage:.2f}%)")
print(f"On-Time Arrivals: {on_time_arrivals} ({arrival_on_time_percentage:.2f}%)")

# Step 4: Create visualization



labels = ['On-Time Departures', 'Delayed Departures']
sizes = [on_time_departures, total_flights - on_time_departures]
colors = ['lightgreen', 'purple']
explode = [0] * len(labels)
explode[1] = 0.1

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, explode = explode, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title(f'On-Time/Delayed Departure Performance for Flight {flight_number} for the past 5 years')
plt.show()

labels = ['On-Time Arrivals', 'Delayed Arrivals']
sizes = [ on_time_arrivals, total_flights - on_time_arrivals]
colors = ['skyblue', 'salmon']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels ,explode = explode, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title(f'On-Time/Delayed Arrival Performance for Flight {flight_number} for the past 5 years')
plt.show()

# Step 4: Create visualization (Bar Chart)
categories = ['On-Time Departures', 'On-Time Arrivals']
percentages = [departure_on_time_percentage, arrival_on_time_percentage]
colors = ['lightgreen','skyblue' ]

plt.figure(figsize=(8, 6))
plt.bar(categories, percentages, color=colors)
plt.xlabel('Performance Category')
plt.ylabel('Percentage')
plt.title(f'On-Time Performance for Flight {flight_number}')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing flight data

# Step 1: Filter the DataFrame for the specific flight number
flight_number = int(input('Enter the flight number:'))
filtered_df = df[df['FL_NUMBER'] == flight_number]

# Step 2: Convert CRS_DEP_TIME and CRS_ARR_TIME to datetime objects
filtered_df['CRS_DEP_TIME'] = pd.to_datetime(filtered_df['CRS_DEP_TIME'], format='%H%M').dt.time
filtered_df['CRS_ARR_TIME'] = pd.to_datetime(filtered_df['CRS_ARR_TIME'], format='%H%M').dt.time

# Step 3: Plot departure delays over time
plt.figure(figsize=(12, 6))
plt.plot(filtered_df['FL_DATE'], filtered_df['DEP_DELAY'], marker='o', linestyle='-', color='blue')
plt.title(f'Departure Delays Over Time for Flight {flight_number}')
plt.xlabel('Date')
plt.ylabel('Departure Delay (Minutes)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 4: Plot arrival delays over time
plt.figure(figsize=(12, 6))
plt.plot(filtered_df['FL_DATE'], filtered_df['ARR_DELAY'], marker='o', linestyle='-', color='orange')
plt.title(f'Arrival Delays Over Time for Flight {flight_number}')
plt.xlabel('Date')
plt.ylabel('Arrival Delay (Minutes)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 5: Group data by day of the week and calculate average delays
filtered_df['FL_DATE'] = pd.to_datetime(filtered_df['FL_DATE'])
filtered_df['Day_of_Week'] = filtered_df['FL_DATE'].dt.day_name()

avg_dep_delay_by_day = filtered_df.groupby('Day_of_Week')['DEP_DELAY'].mean()
avg_arr_delay_by_day = filtered_df.groupby('Day_of_Week')['ARR_DELAY'].mean()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


# Step 6: Plot average departure delays by day of the week
plt.figure(figsize=(8, 6))
avg_dep_delay_by_day.loc[day_order].plot(kind='bar', color='skyblue')
plt.title(f'Average Departure Delay by Day of the Week for Flight {flight_number}')
plt.xlabel('Day of the Week')
plt.ylabel('Average Departure Delay (Minutes)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Step 7: Plot average arrival delays by day of the week
plt.figure(figsize=(8, 6))
avg_arr_delay_by_day.loc[day_order].plot(kind='bar', color='orange')
plt.title(f'Average Arrival Delay by Day of the Week for Flight {flight_number}')
plt.xlabel('Day of the Week')
plt.ylabel('Average Arrival Delay (Minutes)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
